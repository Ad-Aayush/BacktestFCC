```python
import os
import pandas as pd
import json 
```

```python
os.chdir('BacktestingData')
```

```python
tickerToName = {}
```

```python
for file in os.listdir('.'):
    print(file)
    if file.endswith('.xlsx'):
        # Load the Excel file
        df = pd.read_excel(file, header=3)
        
        # Drop rows where 'CD_Bloomberg Code' is NaN
        df = df[pd.notna(df['CD_Bloomberg Code'])]
        # print(df)
        # Convert the 'CD_Bloomberg Code' and 'Company Name' columns to dictionaries and update the main dictionary
        tickerToName.update(pd.Series(df['Company Name'].values, index=df['CD_Bloomberg Code']).to_dict())

print(tickerToName)
```

```python
with open('mydata.json', 'w') as f:
    json.dump(tickerToName, f)
```
